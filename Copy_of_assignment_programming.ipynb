4. The **sample covariance** of two vectors $x=(x_1,...,x_N)$ and $y=(y_1,...,y_N)$ is defined as
$$
cov(x,y) = \dfrac{(x_1 - \bar{x})(y_1-\bar{y}) + (x_2 - \bar{x})(y_2-\bar{y}) + ... + (x_N - \bar{x})(y_{N}-\bar{y})}{N-1}

$$

The intuition of this quantity is that it looks at the pairs $(x_i, y_i)$ and compares them to the means $(\bar{x},\bar{y})$ to determine whether $x$ and $y$ tend to co-vary in the same direction relative to their means: If the values of $x$ and $y$ are typically both above or below the mean values of $x$ and $y$, then $x$ and $y$ will have a positive covariance, but if $x$ is typically above the mean of $x$ when $y$ is typically below the mean of $y$ or vice versa, then they will have a negative covariance. 6. The **sample correlation coefficient** of two vectors $x$ and $y$ is defined as
$$
r_{xy} = \dfrac{cov(x,y)}{s_x s_y}
$$
Use your functions to create a new function that compute this quantity. The intuition of this quantity is that it is like the covariance, but normalized so that its values like between -1 and 1: perfect negative linear association between the variables at -1, no association at 0, and perfect positive linear association between the variables at 1.

```python
import numpy as np
import math as math
np.random.seed(100) # Set the seed for the random number generator
rho, sigma_x, sigma_y = -.4, 3, 2 # Variance-Covariance Parameters
vcv = np.array([[sigma_x**2, rho*sigma_x*sigma_y],
                [rho*sigma_x*sigma_y,sigma_y**2]]) # VCV Matrix
mu = np.array([-1,2]) # Population averages
sample = np.random.multivariate_normal(mu,vcv,200) # Multivariate normal draws
x = sample[:,0]
y = sample[:,1]

#############################################################################

# Computing the sample average (mean)
def mean(x):
  m = np.sum(x) / len(x) # Sum of values in vector divided by size of vector
  return m
print('Sample mean:', mean(x), '\n')

# Computing the sample standard deviation
def ssd(x):
  mxv = (x - mean(x)) ** 2 # subtracting the mean from all values in x and squaring those results
  s = np.sqrt(sum(mxv) / (len(mxv) - 1)) # summing resulting values, dividing by size - 1, and taking square root
  return s
print('SSD:', ssd(x), '\n')

#TODO
# Computing Z-scores
def z_scores(x):
  z = (x - mean(x)) / ssd(x) # Dividing each difference between the value and the mean in the vector by the standard deviation
  return z
print('Z-scores:', z_scores(x)[1:10], '\n')

# Computing sample covariance
def covariance(x, y):
  mxv = x - mean(x) # computing modified x vector
  myv = y - mean(y) # computing modified y vector
  cv = (np.sum(mxv * myv)) / (len(mxv) - 1) # dividing the sum of the products of the modified vectors by the length - 1 (assuming both vectors are the same length)
  return cv
print('covariance:', covariance(x, y), '\n')

# Computing sample correlation coefficient
def scc(x, y):
  r = covariance(x, y) / (ssd(x) * ssd(y)) # dividing the covariance of x and y by the product of the SSDs of x and y
  return r
print('sample correlation coefficient:', scc(x, y), '\n')

```

In this question, you're going to write a function that finds the highest possible value for some function, and returns the value of the function and the maximizing value.

1. Write a function that creates a grid from some value $a$ to a second value $b$ with $N$ steps, computes the value of a function $f$ on that grid, and then finds the maximizers of the function. Have it return a dictionary with the maximum value and the maximizers themselves.
2. Use your function to maximize $f(x) = 100 - 2x^2 + 3x$ for values of $a=-1$, $b=1$, and for values of $N=3$, $N=10$, $N=100$, $N=1000$ and $N=5000$.
3. The true maximizer of this function is $.75$ and the true maximum value is $101.125$. Is that what you got in the previous step? Why not? How does the quality of the maximization depend on $N$? How does your computed answer change as $N$ gets larger?

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): # creating the function to be evaluated
  y = 100 - 2 * x**2 + 3 * x
  return y

def maximize(a, b, f, n):
  x = np.linspace(a, b, n) # creating the grid
  y = f(x) # evaluating the function on the grid

  maximum = max(y) # maximum is the max value the function output
  max_index = np.where(maximum == y) # finding index in the grid where max is located
  maximizer = x[max_index] # finding x-value at max index

  return {'Maximum': maximum, 'Maximizer': maximizer} # creating dictionary

a = -1
b = 1

maximize(a, b, f, 3)
maximize(a, b, f, 10)
maximize(a, b, f, 100)
maximize(a, b, f, 1000)
maximize(a, b, f, 5000)
```
